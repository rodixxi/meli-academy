# TODO: Corregir valores en dolares

```python
first_quartile, third_quartile = df['price'].quantile([0.25, 0.75])
iqr = third_quartile - first_quartile
min = first_quartile - (1.5 * iqr)
max = third_quartile + (1.5 * iqr)
df = df[(df["price"] < max) & (df["price"] > min)]
```

```python
plt.figure(figsize=(16, 6))
sns.set(style="darkgrid")
ax = sns.distplot(df.price)
plt.show()
```

```python
g0, g1, g2, g3 = df['price'].quantile([0.25, 0.45, 0.55, 0.75])
```

```python
import gmplot
gmap = gmplot.GoogleMapPlotter(-31.4168344, -64.1836041, 14)
gmap.scatter(df[df.price <= g0].lat.dropna().tolist(),df[df.price <= g0].lon.dropna().tolist(), '#00FF00', size = 20, marker = False)
gmap.scatter(df[(df.price > g0) & (df.price <= g1)].lat.dropna().tolist(),df[(df.price > g0) & (df.price <= g1)].lon.dropna().tolist(), '#88FF00', size = 20, marker = False) 
gmap.scatter(df[(df.price > g1) & (df.price <= g2)].lat.dropna().tolist(),df[(df.price > g1) & (df.price <= g2)].lon.dropna().tolist(), '#FFFF00', size = 20, marker = False) 
gmap.scatter(df[(df.price > g2) & (df.price <= g3)].lat.dropna().tolist(),df[(df.price > g2) & (df.price <= g3)].lon.dropna().tolist(), '#FF8800', size = 20, marker = False) 
gmap.scatter(df[df.price > g3].lat.dropna().tolist(),df[df.price > g3].lon.dropna().tolist(), '#FF0000', size = 20, marker = False) 
gmap.draw("cba.html") 
```

```python
table = pd.pivot_table(df, values='price', index=['l4'], aggfunc=np.mean)
mas_caros = table.sort_values('price', ascending=False).iloc[:3]
mas_caros
```

```python
mas_baratos = table.sort_values('price', ascending=True).iloc[:3]
mas_baratos
```

```python
barrios_caros = mas_caros.index 
barrios_baratos = mas_baratos.index
```

```python
df = df.dropna(subset=['surface_total'])
```

```python
plt.figure(figsize=(16, 6))
sns.set(style="darkgrid")
g = sns.distplot(df[df.l4.isin(barrios_baratos)].surface_total, label="Barrios baratos")
sns.distplot(df[df.l4.isin(barrios_caros)].surface_total, label="Barrios caros")
g.set_xlim(0, 400)
plt.legend();
```
